# Aplicativos windows store atualizado em 16 out 2021

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('windows_store.csv')
df.head()
```

# Quais são os tópicos da tabela?

```python
for i in df.columns:
    print(i)
```

# Quais os 10 primeiros aplicativos da Windows Store? 

```python
df = df.dropna(how="any", axis= 0)
df['Date of Release'] = pd.to_datetime(df['Date of Release'], errors="coerce")
primeiros_apps = df.sort_values('Date of Release',ascending=True)
primeiros_apps['Date of Release'] = df['Date of Release'].dt.strftime('%d/%m/%Y')
primeiros_apps = primeiros_apps[['Name','Category','Date of Release']]
primeiros_apps = primeiros_apps.head(n=10)
primeiros_apps.index = range(1,11)
primeiros_apps
```

# Quais os 10 últimos aplicativos da Windows Store? 

```python
ultimos_apps = df.sort_values('Date of Release',ascending=False)
ultimos_apps = ultimos_apps[['Name','Category','Date of Release']]
ultimos_apps = ultimos_apps.head(n=10)
ultimos_apps.index = range(1,11)
ultimos_apps
```

# Qual a porcentagem de categorias na Windows Store?

```python
categories = df.groupby(['Category']).count()
categories = categories[['Name']]
categories = categories.rename(columns={'Name':'Count'})
categories['Percentage'] = (categories['Count']/ categories['Count'].sum())*100
others = categories['Percentage']
categories = categories.sort_values('Percentage', ascending=False)
categories = categories.reset_index()
categories.index = categories.index + 1
categories
```

# Maiores quantidade de aplicativos por Categorias

```python
categories = categories[categories['Percentage']>=3]
categories
```

```python
y = categories['Category']
x = categories['Percentage']
plt.figure(figsize = (16,8))
sns.barplot(x, y, orient = "h", palette = "terrain")
plt.ylabel('Categorias')             
plt.xlabel("\nPercentual de aplicativos (%)")
plt.title('\nCategorias que mais aparecem na Store\n')
plt.show()
```

# Aplicativos mais caros (A moeda é a Rupia, uma moeda indiana)

```python
expensive = df[['Name','Price']].copy(deep=True) #Evita erros SettingWithCopyWarning: 
#A value is trying to be set on a copy of a slice from a DataFrame.Try using .loc[row_index,col_indexer] = value instead
```

```python
expensive['Price'] = expensive['Price'].str.replace(r'₹', '')
expensive['Price'] = expensive['Price'].str.replace(r'Free', '0')
expensive['Price'] = expensive['Price'].str.replace(r'Full price was', '')
expensive['Price'] = expensive['Price'].str.replace(r',', '')
```

```python
expensive['Price'] = pd.to_numeric(expensive['Price'], errors='coerce')
```

```python
expensive = expensive.sort_values('Price', ascending=False)
expensive.dropna(subset = ["Price"], inplace=True)
expensive10 = expensive.head(n=10)
expensive10.index = range(1,11)
expensive10
```
